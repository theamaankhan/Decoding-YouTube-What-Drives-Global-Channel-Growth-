```python
category_subscriber_mean = df.groupby("category")["subscribers"].mean()
category_subscriber_mean
```

```python
category_subscriber_mean.sort_values(ascending = False, inplace= True)
```

```python
# plotting a bar chart "Mean subscription vs Category"

category_subscriber_mean.plot(kind ="bar")
plt.title("Mean Subscription by category")
plt.xlabel("Category")
plt.ylabel("Mean Subscribers")
plt.show()
```

```python
channel_type_counts =df["channel_type"].value_counts()
channel_type_counts
```

```python
import matplotlib.pyplot as plt
top_categories = channel_type_counts.sort_values(ascending=False).head(8)
other_sum = channel_type_counts.sum() - top_categories.sum()
top_categories["Others"] = other_sum
plt.figure(figsize=(9, 9))
wedges, texts, autotexts = plt.pie(
    top_categories,
    labels=top_categories.index,
    autopct="%1.1f%%",
    startangle=120,
    colors=plt.cm.Paired.colors,
    pctdistance=0.8,
    labeldistance=1.05,
    wedgeprops={'edgecolor': 'white', 'linewidth': 1.5}
)
plt.setp(autotexts, size=10, weight="bold", color="white")
plt.setp(texts, size=11, weight="medium")
plt.title("📊 YouTube Channel Type Distribution", fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.figtext(
    0.5, 0.02,
    "🎬 Entertainment and Music dominate YouTube — together they form over half of all channels.",
    ha="center", fontsize=11, color="dimgray"
)

plt.show()
```

```python
plt.figure(figsize=(9, 6))
sns.set(style="whitegrid")
plt.scatter(
    df["subscribers"],
    df["video views"],
    alpha=0.6,
    s=60,
    c=df["subscribers"],
    cmap="coolwarm",
    edgecolor="white", linewidth=0.7
)

plt.title("📈 Subscribers vs Video Views", fontsize=16, fontweight='bold', pad=15)
plt.xlabel("Subscribers (in numbers)", fontsize=12)
plt.ylabel("Video Views (in total)", fontsize=12)
plt.ticklabel_format(style='plain', axis='x')
plt.ticklabel_format(style='plain', axis='y')
cbar = plt.colorbar()
cbar.set_label("Number of Subscribers", rotation=270, labelpad=15)
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.figtext(
    0.5, -0.05,
    "💡 Insight: A positive correlation is visible — channels with higher subscribers tend to have more total video views.",
    ha="center", fontsize=11, color="dimgray"
)

plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.regplot(data=df, x='subscribers', y='video views', scatter_kws={'alpha':0.6, 's':50}, line_kws={'color':'red', 'lw':2})
plt.xscale('log')  # optional to reduce skew
plt.yscale('log')  # optional
plt.xlabel('Subscribers')
plt.ylabel('Video Views')
plt.title('Subscribers vs Video Views (regression)')
plt.tight_layout()
plt.show()
```

## Conclusions & Key Results

- The dataset contains **995 channels** and **28 columns** of metadata. The sample is skewed toward large entertainment channels.
- **Top category**: Entertainment (241 channels among the largest groups). Top countries by channel count are **United States (313)**, **India (168)** and **Brazil (62)**.
- Subscriber statistics are heavily right-skewed (mean ≈ **22.98M**, median = **17.7M**, max = **245M**). Views are similarly skewed (mean ≈ **11.04B**, median ≈ **7.76B**, max ≈ **228B**).
- There is a **strong positive correlation (~0.75)** between subscribers and total video views: larger subscriber bases tend to generate more views, but notable outliers show that individual viral videos can lead to very high view counts even with lower subscriber counts.
- **Actionable insights**:
  - For reach: focus on Entertainment and Music categories.
  - For engagement & conversions: Education/How-To channels are promising.
  - For creators: consistent posting cadence and regionally tailored content accelerate growth.
